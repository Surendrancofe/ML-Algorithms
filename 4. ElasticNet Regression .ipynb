## ElasticNet Regression

```python
import pandas as pd
```

```python
data = pd.read_csv("Cotton.csv")
```

```python
data.info()
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python
df = pd.DataFrame(data)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_features = ["Region", "Soil_Type", "Fertilizer_Used", "Irrigation_Used", "Weather_Condition"]
```

```python
label_encoders = {}
```

```python
for feature in categorical_features:
    le = LabelEncoder()
    data[feature] = le.fit_transform(df[feature])
    label_encoders[feature] = le
```

```python
X = data.drop(columns=["Yield_tons_per_hectare","Crop"])
y = data["Yield_tons_per_hectare"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
model = ElasticNet()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
MSE = mean_squared_error(y_test, predictions)
```

```python
print(f"Mean Squared Error: {MSE}")
```

```python
MAE = mean_absolute_error(y_test, predictions)
```

```python
print(f"Mean Absolute Error: {MAE}")
```

```python
RSquared = r2_score(y_test, predictions)
```

```python
print(f"RSquared Score: {RSquared}")
```

# Hyperparameter Tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "alpha": [0.001, 0.01, 0.1, 1, 10, 100],
    "l1_ratio": [0.1, 0.3, 0.5, 0.7, 0.9, 1.0]
}
```

```python
grid_search = GridSearchCV(model, param_grid, cv=5, scoring="neg_mean_squared_error")
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
```

```python
best_model.fit(X_train, y_train)
```

```python
predictions = best_model.predict(X_test)
```

```python
MSE = mean_squared_error(y_test, predictions)
```

```python
print(f"Mean Squared Error: {MSE}")
```

```python
MAE = mean_absolute_error(y_test, predictions)
```

```python
print(f"Mean Absolute Error: {MAE}")
```

```python
RSquared = r2_score(y_test, predictions)
```

```python
print(f"RSquared Score: {RSquared}")
```

# Hyperparameter Tuning using Optuna

```python
import optuna
```

```python
#Define the Optuna Objective Function
def objective(trial):
    alpha = trial.suggest_float('alpha', 1e-3, 1e3)  # Regularization strength
    l1_ratio = trial.suggest_float('l1_ratio', 0.1, 1.0)  # Mix of L1 and L2

    # Define and train the model
    model = ElasticNet(alpha=alpha, l1_ratio=l1_ratio, max_iter=10000)
    model.fit(X_train, y_train)

    # Predict and evaluate using Mean Squared Error
    y_pred = model.predict(X_test)
    MSE = mean_squared_error(y_test, y_pred)
    return MSE
```

```python
# Create an Optuna study and optimize it
study = optuna.create_study(direction='minimize')  # Minimize MSE
study.optimize(objective, n_trials=500)
```

```python
study.best_params
```

```python
study.best_value
```

```python
best_params = study.best_params
```

```python
best_model = ElasticNet(**best_params)
```

```python
best_model.fit(X_train, y_train)
```

```python
predictions = best_model.predict(X_test)
```

```python
MSE = mean_squared_error(y_test, predictions)
```

```python
print(f"Mean Squared Error: {MSE}")
```

```python
MAE = mean_absolute_error(y_test, predictions)
```

```python
print(f"Mean Absolute Error: {MAE}")
```

```python
RSquared = r2_score(y_test, predictions)
```

```python
print(f"RSquared Score: {RSquared}")
```
