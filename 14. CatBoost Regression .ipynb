# CatBoost Regression

```python
import pandas as pd
```

```python
data = pd.read_csv("Cotton.csv")
```

```python
data.info()
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python
df = pd.DataFrame(data)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_features = ["Region", "Soil_Type", "Fertilizer_Used", "Irrigation_Used", "Weather_Condition"]
```

```python
label_encoders = {}
```

```python
for feature in categorical_features:
    le = LabelEncoder()
    data[feature] = le.fit_transform(df[feature])
    label_encoders[feature] = le
```

```python
X = data.drop(columns=["Yield_tons_per_hectare","Crop"])
y = data["Yield_tons_per_hectare"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from catboost import CatBoostRegressor
```

```python
model = CatBoostRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
MSE = mean_squared_error(y_test, predictions)
```

```python
print(f"Mean Squared Error: {MSE}")
```

```python
MAE = mean_absolute_error(y_test, predictions)
```

```python
print(f"Mean Absolute Error: {MAE}")
```

```python
RSquared = r2_score(y_test, predictions)
```

```python
print(f"RSquared Score: {RSquared}")
```

# Hyperparameter Tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'iterations': [100, 300, 500],  
    'learning_rate': [0.01, 0.05, 0.1],  
    'depth': [4, 6, 8, 10],  
    'l2_leaf_reg': [1, 3, 5, 10],  
    'bagging_temperature': [0.1, 1, 5],  
    'random_strength': [0.1, 1, 5],  
    'border_count': [32, 64, 128]
}
```

```python
grid_search = GridSearchCV(model, param_grid, cv=5, scoring="neg_mean_squared_error")
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
```

```python
best_model.fit(X_train, y_train)
```

```python
predictions = best_model.predict(X_test)
```

```python
MSE = mean_squared_error(y_test, predictions)
```

```python
print(f"Mean Squared Error: {MSE}")
```

```python
MAE = mean_absolute_error(y_test, predictions)
```

```python
RSquared = r2_score(y_test, predictions)
```

```python
print(f"RSquared Score: {RSquared}")
```

# Hyperparameter Tuning using Optuna

```python
import optuna
```

```python
# Define the Optuna objective function
def objective(trial):
    params = {
        'iterations': trial.suggest_int('iterations', 100, 1000),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'depth': trial.suggest_int('depth', 4, 10),
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1, 10),
        'bagging_temperature': trial.suggest_float('bagging_temperature', 0.1, 5),
        'random_strength': trial.suggest_float('random_strength', 0.1, 5),
        'border_count': trial.suggest_int('border_count', 32, 255)
    }

    # Define and train the model
    model = CatBoostRegressor(loss_function='RMSE', random_state=42, verbose=0, **params)
    model.fit(X_train, y_train)

    # Predict and evaluate using Mean Squared Error
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    return mse
```

```python
# Create an Optuna study and optimize it
study = optuna.create_study(direction='minimize')  # Minimize MSE
study.optimize(objective, n_trials=50)
```

```python
study.best_params
```

```python
study.best_value
```

```python
best_params = study.best_params
```

```python
best_model = CatBoostRegressor(**best_params)
```

```python
best_model.fit(X_train, y_train)
```

```python
predictions = best_model.predict(X_test)
```

```python
MSE = mean_squared_error(y_test, predictions)
```

```python
print(f"Mean Squared Error: {MSE}")
```

```python
MAE = mean_absolute_error(y_test, predictions)
```

```python
print(f"Mean Absolute Error: {MAE}")
```

```python
RSquared = r2_score(y_test, predictions)
```

```python
print(f"RSquared Score: {RSquared}")
```
