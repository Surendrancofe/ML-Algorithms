## Decision Tree Regression

```python
import pandas as pd
```

```python
data = pd.read_csv("Cotton.csv")
```

```python
data.info()
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python
df = pd.DataFrame(data)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_features = ["Region", "Soil_Type", "Fertilizer_Used", "Irrigation_Used", "Weather_Condition"]
```

```python
label_encoders = {}
```

```python
for feature in categorical_features:
    le = LabelEncoder()
    data[feature] = le.fit_transform(df[feature])
    label_encoders[feature] = le
```

```python
X = data.drop(columns=["Yield_tons_per_hectare","Crop"])
y = data["Yield_tons_per_hectare"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
MSE = mean_squared_error(y_test, predictions)
```

```python
print(f"Mean Squared Error: {MSE}")
```

```python
MAE = mean_absolute_error(y_test, predictions)
```

```python
print(f"Mean Absolute Error: {MAE}")
```

```python
RSquared = r2_score(y_test, predictions)
```

```python
print(f"RSquared Score: {RSquared}")
```

# Hyperparameter Tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "max_depth": [3, 5, 10, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ['auto', 'sqrt', 'log2', None]
}
```

```python
grid_search = GridSearchCV(model, param_grid, cv=5, scoring="neg_mean_squared_error")
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
```

```python
best_model.fit(X_train, y_train)
```

```python
predictions = best_model.predict(X_test)
```

```python
MSE = mean_squared_error(y_test, predictions)
```

```python
print(f"Mean Squared Error: {MSE}")
```

```python
MAE = mean_absolute_error(y_test, predictions)
```

```python
print(f"Mean Absolute Error: {MAE}")
```

```python
RSquared = r2_score(y_test, predictions)
```

```python
print(f"RSquared Score: {RSquared}")
```

# Hyperparameter Tuning with Optuna

```python
import optuna
```

```python
def objective(trial):
    max_depth = trial.suggest_int('max_depth', 2, 20)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 20)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 10)
    max_features = trial.suggest_categorical('max_features', ['sqrt', 'log2', None])
    #Define and Train the Model
    model = DecisionTreeRegressor(
        max_depth = max_depth,
        min_samples_split = min_samples_split,
        min_samples_leaf = min_samples_leaf,
        max_features = max_features,
        random_state = 42
    )
    model.fit(X_train, y_train)
    #Predict and Evaluate using Mean Squared Error
    predictions = model.predict(X_test)
    MSE = mean_squared_error(y_test, predictions)
    return MSE
```

```python
#Create an Optuna Study and Optimize it
study = optuna.create_study(direction = 'minimize')
study.optimize(objective, n_trials = 100)
```

```python
study.best_params
```

```python
study.best_value
```

```python
best_params = study.best_params
```

```python
best_model = DecisionTreeRegressor(**best_params)
```

```python
best_model.fit(X_train, y_train)
```

```python
predictions = best_model.predict(X_test)
```

```python
MSE = mean_squared_error(y_test, predictions)
```

```python
print(f"Mean Squared Error: {MSE}")
```

```python
MAE = mean_absolute_error(y_test, predictions)
```

```python
print(f"Mean Absolute Error: {MAE}")
```

```python
RSquared = r2_score(y_test, predictions)
```

```python
print(f"RSquared Score: {RSquared}")
```
